```python
# Apriori algorithm

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\HP\Lab_Works_SEM5\a.csv', sep='\t', header=None)
df
```

```python
len(df)
```

```python
z=[]
for i in range(len(df)):
    z.append(df.values[i,0].split(","))
z
```

```python
from apyori import apriori
rules = apriori(transactions = z, min_support = 0.001, min_confidence = 0.2)
r = list(rules)
r
```

```python
for result in r:
    items = [item for item in result.items]
    support = result.support
    confidence = result.ordered_statistics[0].confidence
    lift = result.ordered_statistics[0].lift
    print(f"Items: {items} Support: {support} Confidence: {confidence} Lift: {lift}")
```
